# Problem 3: Data Cleaning

```python
# Identify columns with more than 80% missing values for informational purposes
missing_percent = df_AHDB.isnull().mean()
cols_over_80pct_missing = missing_percent[missing_percent > 0.8].index.tolist()
print("Columns with more than 80% missing values:")
print(cols_over_80pct_missing)

# Identify columns with a single unique value
single_unique_cols = [col for col in df_AHDB.columns if df_AHDB[col].nunique(dropna=False) == 1]
print("Columns with a single unique value:")
print(single_unique_cols)

# Remove columns with a single unique value from the DataFrame
df_AHDB = df_AHDB.drop(columns=single_unique_cols)

```

## Insights

Missing Data Strategy (Alley example):

The Alley column contains a large number of missing values.

Since most homes do not have an alley, it is likely that NaN actually means "No Alley Access".

✅ Decision: Instead of dropping the column, replace NaN with a new category like "NoAlley" to retain useful information while avoiding data loss.

Columns with Over 80% Missing Values:

Columns like Alley, PoolQC, Fence, and MiscFeature have more than 80% missing data.

✅ Recommendation: These columns may be dropped unless they are known to carry high predictive value or domain-specific importance.

Keeping them may introduce noise or bias during modeling.

Columns with a Single Unique Value:

No columns were found to have only one unique value in this dataset.

If such columns existed, they would be removed as they provide no variability and no predictive power.

# Problem 4: Aggregation and Grouping

```python
# 4.1 Median sale price of houses by building type
median_saleprice_by_bldgtype = df_AHDB.groupby('BldgType')['SalePrice'].median()
print("Median sale price of houses by building type:")
print(median_saleprice_by_bldgtype)

# 4.2 Year with the highest number of houses sold
yearly_sales_counts = df_AHDB['YrSold'].value_counts()
year_with_most_sales = yearly_sales_counts.idxmax()
num_sales = yearly_sales_counts.max()
print(f"Year with the highest number of houses sold: {year_with_most_sales} ({num_sales} sales)")


# 4.3 Average sale price by neighborhood and building type
avg_saleprice_by_neigh_bldg = df_AHDB.groupby(['Neighborhood', 'BldgType'])['SalePrice'].mean()
print("Average sale price by neighborhood and building type:")
print(avg_saleprice_by_neigh_bldg)

# 4.4 Year with the highest and lowest number of sales
yearly_sales_counts = df_AHDB['YrSold'].value_counts()
year_with_most_sales = yearly_sales_counts.idxmax()
num_most_sales = yearly_sales_counts.max()
year_with_least_sales = yearly_sales_counts.idxmin()
num_least_sales = yearly_sales_counts.min()
print(f"Year with the highest number of sales: {year_with_most_sales} ({num_most_sales} sales)")
print(f"Year with the lowest number of sales: {year_with_least_sales} ({num_least_sales} sales)")
```

## Insights

✅ Key Insights:
4.1 Median Sale Price by Building Type:
The most expensive building type, based on median SalePrice, is TwnhsE (Townhouse End Unit), with a median of $172,200.

```python

# 6.1 Create a correlation matrix of the numerical columns
corr_matrix = df.corr(numeric_only=True)
print(corr_matrix)

# Visualize the correlation matrix using a heatmap
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20, 18))  # רוחב וגובה מוגדלים
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True, annot_kws={"size": 8})
plt.title("Correlation Matrix Heatmap", fontsize=16)
plt.xticks(rotation=45, ha='right')  # סיבוב טקסט בציר X
plt.yticks(rotation=0)               # השארת טקסט בציר Y ישר
plt.tight_layout()                   # סידור אוטומטי למניעת חפיפות
plt.show()

# 6.2 Visualize the trend of average sale price over the years


avg_price_per_year = df.groupby('YrSold')['SalePrice'].mean()

plt.figure(figsize=(8, 6))
sns.lineplot(x=avg_price_per_year.index, y=avg_price_per_year.values, marker='o')
plt.xlabel('Year Sold')
plt.ylabel('Average Sale Price')
plt.title('Average Sale Price Trend Over Years')
plt.grid(True)
plt.show()


# 6.3 Scatter plot for GrLivArea vs. SalePrice
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='GrLivArea', y='SalePrice', alpha=0.6)
plt.xlabel('Above Grade Living Area (sq ft)')
plt.ylabel('Sale Price')
plt.title('GrLivArea vs. SalePrice')
plt.grid(True)
plt.show()


# 6.4 Boxplot for SalePrice in different neighborhoods
plt.figure(figsize=(16, 8))
sns.boxplot(data=df, x='Neighborhood', y='SalePrice')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Neighborhood')
plt.ylabel('Sale Price')
plt.title('Sale Price Distribution by Neighborhood')
plt.tight_layout()
plt.show()

```

The variable OverallQual (overall material and finish quality) shows a strong positive correlation with SalePrice.

Some variables (e.g., BsmtUnfSF, LowQualFinSF) show near-zero or negative correlation, suggesting they have little predictive value for price.

📈 6.2 Sale Price Trend Over Years:
The line plot shows that average home prices remain relatively stable over the years, with minor fluctuations.

There are no significant uptrends or downtrends, implying that the market was fairly consistent during the period covered.

🏠 6.3 GrLivArea vs. SalePrice:
